# Makine Öğrenmesi Nedir?

---



Makine öğrenmesi, bilgisayar sistemlerinin önceden belirlenmiş veri ve algoritmalar kullanarak, belirli bir görevi otomatik olarak öğrenmesini sağlayan bir teknolojidir. Bu teknoloji, verilerin yüksek hacimlerinde analiz yaparak, belirli bir hedefe ulaşmak için yeni bilgiler üretir. Bu süreçte, makine öğrenmesi algoritmaları, girdi verilerindeki özellikleri tanımlar ve bu özelliklerin bir modelde nasıl kullanılacağına karar verir.

**Makine Öğrenmesi Nasıl Çalışır?**

Makine öğrenmesi, veri setleri üzerinde bir dizi matematiksel işlem gerçekleştirir. Bu işlemler, önceden belirlenmiş bir algoritma tarafından yönetilir. Algoritma, veri setinin yapısına bağlı olarak özelleştirilir. Makine öğrenmesi algoritmaları, veri setlerindeki örüntüleri ve ilişkileri analiz eder. Sonuç olarak, bu algoritmalar verileri sınıflandırmak, öngörmek veya çıkarımlarda bulunmak için öğrenirler.
Makine öğrenmesi, genellikle üç farklı öğrenme şekliyle gerçekleştirilir:

**Denetimli Öğrenme**

Denetimli öğrenme, bir makine öğrenmesi tekniği türüdür ve belirli bir çıktı veya hedef için belirli bir girdiye dayalı olarak bir modelin eğitildiği bir öğrenme sürecidir. Bu yöntem, belirli bir veri kümesindeki desenleri öğrenerek, gelecekteki yeni veriler için doğru tahminler yapmak için kullanılabilir. Öğrenme verileri, girdiler ve çıktılar olarak adlandırılan özellikleri içeren bir veri kümesinden oluşur. Denetimli öğrenme, sınıflandırma ve regresyon problemlerini çözmek için kullanılır. Sınıflandırma, örneğin bir veri noktasının belirli bir sınıfa ait olup olmadığını tahmin etmek için kullanılırken, regresyon, örneğin bir veri noktasının belirli bir sayısal değerini tahmin etmek için kullanılır.

```python
# Scikit-learn kütüphanesini kullanarak bir kümeleme modeli oluşturun
from sklearn.cluster import KMeans
from sklearn import datasets

# Veri kümesini yükleyin
iris = datasets.load_iris()
X = iris.data

# Modeli oluşturun ve uygulayın
model = KMeans(n_clusters=3, random_state=0)
model.fit(X)

# Tahmin sonuçlarını görselleştirin
import matplotlib.pyplot as plt

plt.scatter(X[:, 0], X[:, 1], c=model.labels_)
plt.xlabel('Sepal length')
plt.ylabel('Sepal width')
plt.show()
```

**Denetimsiz Öğrenme**

Denetimsiz öğrenme, bir makine öğrenmesi modelinin, etiketlenmemiş veriler üzerinden öğrenmesini sağlayan bir öğrenme şeklidir. Etiketlenmemiş veriler, veri setindeki herhangi bir özelliğe sahip verilerdir. Model, bu verilere dayalı olarak veri setindeki örüntüleri ve yapıları keşfetmek için kullanılabilir.

Denetimsiz öğrenme, veriler arasındaki ilişkileri keşfetmek ve özelliklerini tanımlamak için kullanılır. Bu özellikler, daha sonra verileri sınıflandırmak veya etiketlemek için kullanılabilir. Denetimsiz öğrenme, bir makine öğrenmesi modelinin verilerdeki karmaşıklığı ele alması ve daha basit yapılar ve özellikler oluşturması için kullanılır.

Denetimsiz öğrenme, özellikle büyük veri kümeleri üzerinde kullanıldığında çok faydalıdır. Bu tür veri kümeleri genellikle etiketlenmemiştir, dolayısıyla denetimsiz öğrenme algoritmaları verilerdeki örüntüleri tanımlamak için kullanılır.

```python
# Scikit-learn kütüphanesini kullanarak bir regresyon modeli oluşturun
from sklearn import datasets, linear_model
import numpy as np
import matplotlib.pyplot as plt

# Veri kümesini yükleyin
diabetes = datasets.load_diabetes()

# Verileri özellikler (X) ve hedef değişken (y) olarak ayırın
X = diabetes.data[:, np.newaxis, 2]
y = diabetes.target

# Modeli oluşturun ve uygunlaştırın
model = linear_model.LinearRegression()
model.fit(X, y)

# Test verileri oluşturun ve tahmin yapın
X_test = np.linspace(min(X), max(X), 500).reshape(-1, 1)
y_pred = model.predict(X_test)

# Tahmin sonuçlarını görselleştirin
plt.scatter(X, y)
plt.plot(X_test, y_pred, color='red', linewidth=3)
plt.show()
```
